## Multinomial Logistic Regression using NumPy

Logistic regression is a supervised machine learning algorithm commonly used in classification task. **Supervised** means each  sample (represented by a float vector) is *labeled*. We have provided some util functions to fetch some generated data for you.

```python
import numpy as np
import numpy.random as random
import time
from util import get_data, plot_data

# Initialize training data
data, label = get_data()
num_samples = data.shape[0]
num_features = data.shape[1]
num_classes = label.shape[1]

print 'Shapes: data {}, label {}'.format(data.shape, label.shape)
print '# samples: {}'.format(num_samples)
print '# features: {}'.format(num_features)
print '# classes: {}'.format(num_classes)
```

In our example, the training set contains 10000 samples, each has 500 features, so the training set is represented by a 10000x500 matrix (`data`). Our goal is to classify each samples into 5 classes. Therefore, each sample comes with a label encoded in *one-of-many* format. For example, say sample 0 is of class 1, then the corresponding label vector for sample 0 will be `[0, 1, 0, 0, 0]`. Therefore, label could be represented by a 10000x5 matrix `label` as follows:

```python
print 'Data matrix:'
print data
print 'Label matrix:'
print label
```

The function `plot_data` visualizes what the data looks like. It projects each 500-D samples onto their first two dimensions, so each sample is a point in the figure and its color identifies its class. 

Although this is a very simple example, its setup is general. For example, in an *Optical Character Recognition (OCR)* task (e.g. recognizing MNIST handwriting digit), you could think about the blue points being the samples of digit "0" and red points digit "1" and so on. Our goal is to correctly classify some *unlabeled* samples (test data) into the most likely classes.

```python
# Visualize what the ground truth data looks like
print 'Ground truth'
plot_data(data, label)
```

The model we use here is *Multinomial Logistic Regression*, or *Softmax Regression*. Logistic regression deals with binary classes (either 0 or 1) only, whereas *multinomial logistic regression* allows classification into multiple exclusive classes. The model contains a weight matrix `w` that projects from feature space to class space (so its dimension is `num_features` $\times$ `num_classes`). It first computes the probability of each class given the samples:
$$p(y^{(i)}=j \mid x^{(i)};w)= \frac{\exp(w_j^Tx^{(i)})}{\sum_j\exp(w_j^Tx^{(i)})}$$
The following code implements the above equation in the `predict` function.

```python
# predict the class using logistic regression
def predict(w, x):
    #===========================================================#
    #                    Your code starts here                  #
    #===========================================================#
    # please try compute the estimated probability of each sample
    a = np.exp(np.dot(x, w))
    a_sum = np.sum(a, axis=1, keepdims=True)
    prob = a / a_sum
    #===========================================================#
    #                    Your code ends here                    #
    #===========================================================#
    return prob
```

We can use the above `predict` function to visualize random guesses as follows; you can see they are completely different from the distribution of correct labels.

```python
# Initialize training weight
weight = random.randn(num_features, num_classes)

# Visualize the initial guess
print 'Initial Guess'
plot_data(data, predict(weight, data))
```

Multinomial logistic regression then tries to minimize the *cross entropy* between the `predict` class probability and the correct probability. More specifically, it tries to minimize the following cost function:
$$J(w)=-\frac{1}{m}\left[\sum_{i=1}^{m}\sum_{j=1}^{k}1\{y^{(i)}=j\}\log (p(y^{(i)}=j \mid x^{(i)};w))\right]$$
, where $m$ is number of samples, $1\{...\}$ is the indicator function which equals to 1 if the condition holds, and 0 otherwise.

The loss is computed in the following code. Note that our label (the correct probability) has already been transformed into *one-of-many* form, so the indicator function here is simply an element-wise multiplication of the label matrix and the logarithmic probability.

To minimize the loss, we use **batch gradient descent**. It first computes the gradient of the weight matrix as follows:
$$\nabla_{w}J(w)=-\frac{1}{m}\sum_{i=1}^{m}\left[x^{(i)}(1\{y^{(i)}=j\}-p(y^{(i)}=j \mid x^{(i)};w))\right]$$
, then updates the current model along the gradient direction:
$$w\leftarrow w - \alpha \nabla_{w}J(w)$$
We have implemented the gradient computation and the weight update for you.

```python
# Using gradient descent to fit the correct classes
def train(w, x, loops):
    for i in range(loops):
        prob = predict(w, x)
        if i % 10 == 0:
            #===========================================================#
            #                    Your code starts here                  #
            #===========================================================#
            # please try compute the current loss value
            loss = -np.sum(label * np.log(prob)) / num_samples
            #===========================================================#
            #                    Your code ends here                    #
            #===========================================================#
            print 'Iter {}, training loss {}'.format(i, loss)
        # calculate gradient of weight
        dy = prob - label
        dw = np.dot(data.T, dy) / num_samples
        # update 
        w -= 0.1 * dw
```

Here are training losses for 100 iterations. You can see the loss value keeps dropping.

```python
# Now training it for 100 iterations
start_time = time.time()
train(weight, data, 100)
print 'Training time: {}s'.format(time.time() - start_time)
```

You can visualize the prediction with trained model again. It should now be very close to what the ground truth is.

```python
# Plot the prediction after training. It should show similar image as the ground truth.
plot_data(data, predict(weight, data))
```

## Multinomial Logistic Regression using MXNet NumPy

### Utilize GPU computation with little (or no) numpy syntax change

Such a tiny example take around 9 seconds with 100 iterations. Real world tasks involve billions of samples and many more features and classes. How to efficiently train such a model? One solution is to use GPU. Our tool, `minpy` allows you to use GPU to speed up the algorithm and in the meantime, keep the neat `numpy` syntax almost unchanged.

```python
# All you need to do is replace the `np` namespace with minpy's.
import minpy.numpy as np
import minpy.numpy.random as random
```

```python
# Initialize weight matrix (again)
weight = random.randn(num_features, num_classes)

# Now call the same training function
# Since the namespace is redefined, it will automatically run on GPU.
start_time = time.time()
train(weight, data, 100)
 # You should observe a significant speed up (around 3x) to the previous training time.
print 'Training time: {}s'.format(time.time() - start_time)
```

### Automatic gradient calculation

Writing the back-propagation portion of the code is tedious and extraordinarily error-prone. Inspired by the package **[augograd](https://github.com/HIPS/autograd)**, `minpy` computes gradient automatically. In teh following code, the `grad_and_loss` function takes your defined `train_loss` function and returns a function computes both the loss value and the gradient of weight. The gradient could then be directly used to update the model weight.

**Regularization:** L2-regularization is widely used to limit the power of the model in order to avoid overfitting. The new loss function is as follows:
$$J'(w)=J(w)+\sum_{i}w_i^2$$ and this is implemented as follows.

```python
from minpy.core import grad_and_loss

# Initialize weight matrix (again)
weight = random.randn(num_features, num_classes)

# Using gradient descent to fit the correct classes
def train_loss(w, x):
    #===========================================================#
    #                    Your code starts here                  #
    #===========================================================#
    prob = predict(w, x)
    loss = -np.sum(label * np.log(prob)) / num_samples + np.sum(w * w)
    #===========================================================#
    #                    Your code ends here                    #
    #===========================================================#
    return loss

# Calculate gradient function automatically
grad_function = grad_and_loss(train_loss)

# Now training it for 100 iterations
start_time = time.time()
for i in range(100):
    dw, loss = grad_function(weight, data)
    if i % 10 == 0:
        print 'Iter {}, training loss {}'.format(i, loss)
    weight -= 0.1 * dw
print 'Training time: {}s'.format(time.time() - start_time)
```

```python
# plot the result
# ATTENTION: when calling an external package implemented by numpy, 
# you need to explicitly convert minpy's array to numpy's array
plot_data(data, predict(weight, data).asnumpy())
```

### Graceful fallback when GPU operators are missing

Numpy is a giant library with hundreds of operators. Our supported GPU operators are only a subset of them, it is likely that your code may include functions that are currently missing on GPU side. To solve this problem, `minpy` gracefully adopts the numpy implementation if the operator is missing on GPU side, and handles the memory copies between GPU and CPU transparently.

```python
# First turn on the logging to know what happens under the hood
import logging
logging.getLogger('minpy.array').setLevel(logging.DEBUG)

x = np.zeros((10, 20))

# cosh is currently missing in MXNet's GPU implementation
# So it will fallback to use numpy's CPU implementation,
# but you don't need to worry about the memory copy from GPU -> CPU
y = np.cosh(x)

# log has GPU implementation, so it will copy the array from CPU -> GPU.
# Once again, you don't need to worry about it. It is transparent.
z = np.log(y)

# Turn off the logging
logging.getLogger('minpy.array').setLevel(logging.WARN)
```

### Use predefined MXNet symbol as one operator

Although you start with numpy operators to build your algorithm, more efficient implementations exist that fuse the mostly commonly operators into a more efficient implementation. For example, convolution is very slow in numpy's implementation, while NVIDIA provides very efficient CUDA implementation. Hence, `minpy` allows you to directly use MXNet's symbol as one operator, and you could mix MXNet's declarative programming with Minpy's imperative programming in one piece of code. The following example replaces the computation of:
$$p(y^{(i)}=j \mid x^{(i)};w)= \frac{\exp(w_j^Tx^{(i)})}{\sum_j\exp(w_j^Tx^{(i)})}$$
 with a predefined `SoftmaxOutput` symbol.

After you have gone through the next tutorial about MNIST OCR task. You could try replace the `np.dot` function with the `sym.FullyConnected` symbol. Remember, you can combine multiple symbols into one operator.

```python
import mxnet
import mxnet.symbol as sym
from minpy.core import function

# define softmax symbol (this part will be covered in the later tutorial)
x_shape = (num_samples, num_classes)
label_shape = (num_samples,)
softmax_symbol = sym.SoftmaxOutput(data=sym.Variable('x'), name='softmax', grad_scale=1.0/num_samples)

# convert MXNet symbol into a callable function (with corresponding gradient function)
softmax = function(softmax_symbol, [('x', x_shape), ('softmax_label', label_shape)])

# make softmax_label; MXNet's softmax operator does not use one-of-many label format
softmax_label = np.argmax(label, axis=1)

# Redefine loss function using softmax as one operator
def train_loss(w, x):
    y = np.dot(x, w)
    prob = softmax(x=y, softmax_label=softmax_label)
    loss = -np.sum(label * np.log(prob)) / num_samples
    return loss

# Initialize weight matrix (again)
weight = random.randn(num_features, num_classes)

# Calculate gradient function automatically
grad_function = grad_and_loss(train_loss)

# Now training it for 100 iterations
start_time = time.time()
for i in range(100):
    dw, loss = grad_function(weight, data)
    if i % 10 == 0:
        print 'Iter {}, training loss {}'.format(i, loss)
    weight -= 0.1 * dw
print 'Training time: {}s'.format(time.time() - start_time)
```
